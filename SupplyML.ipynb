```python
def calculateConfidence(x):
    total = x['On-Time']+x['Quality']+x['Quantity Met']
    
    if total == 0:
        return 0.0
    elif total == 1:
        return 0.33
    elif total == 2:
        return 0.66
    else:
        return 1.0
```

```python
df['Confidence'] = df.apply(lambda x: calculateConfidence(x),axis=1)
```

```python
df['Location'].unique()
```

```python
df_prd = df[df["Item Unique Identifier"] == "HD 700Desert TaupeQueen"]
```

# Filtering only 0.66 Confidence rows

```python
df_66 = df_prd[df_prd["Confidence"] == 0.66]
df_66
```

## Dropping irrerlevant columns

```python
df_up66 = df_66.drop(["Short Quantity","Received Qty","Rejected Quantity","Time Promised","Time Received","On-Time","Quality","Quantity Met","Confidence"] , axis=1)
df_up66
```

## Calculating Net Difference as Net Quantity - Expected Quantity

```python
df_up66["NetDiff"] = df_up66["Net Quantity"] - df_up66["Expected Qty"]
df_up66
```

## Calculating mean of Net Difference and group by supplier

```python
df_up66 = df_up66.groupby('Supplier', as_index=False)['NetDiff'].mean()
df_up66
```

## And performing same steps which were performed for 0.66 (above)

```python
df_1 = df_prd[df_prd["Confidence"] == 1.0]
df_1
```

```python
df_up1 = df_1.drop(["Short Quantity","Received Qty","Rejected Quantity","Time Promised","Time Received","On-Time","Quality","Quantity Met","Confidence"] , axis=1)
df_up1
```
